```python
def kdehist(data, title):
    x = data.price
    fig, ax = plt.subplots(2,1,figsize=(20,10),sharex=True)
    sns.histplot(x,ax=ax[0])
    ax[0].set(xlabel=None)
    ax[0].set_title('Histogram + KDE')
    sns.violinplot(x, ax=ax[1])
    ax[1].set(xlabel=None)
    ax[1].set_title('Violin plot')
    fig.suptitle(title, fontsize=16)
    plt.tight_layout(pad=3.0)
    plt.show()
    norm = plt.Normalize(x.min(), x.max())
    sm = plt.cm.ScalarMappable(norm=norm)
    sm.set_array([])
    ax=sns.scatterplot(x="longitude",y="latitude",data=data, alpha=0.4,hue="price",palette="Dark2")
    ax.get_legend().remove()
    s=ax.figure.colorbar(sm)
    s.ax.set_title("Price")
kdehist(df,"Delhi Price(in Million ruppe)")    
```

```python
def corr_func(data,title):
    plt.figure(figsize=(30,25))
    corr_matrix = data.corr(numeric_only=True)
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    fig = sns.heatmap(corr_matrix,mask=mask,cmap='YlGnBu',vmax=1,center=0,fmt='f',square=True,linewidths=.5,annot=True)
    plt.title(title,fontsize=24)
    plt.show() 

corr_func(df,"Delhi Correlation Heatmap")
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("./data/delhi_cleaned.csv")
```

```python
df.head() 
```

```python
num = [i for i in df.columns if df[i].dtype in ["int64", "float64"]]
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = 3
n_rows = (len(numeric_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5*n_cols, 5*n_rows))

for idx, col in enumerate(numeric_cols):
    ax = axes.flatten()[idx]
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(col)

# Hide unused axes
for idx in range(len(numeric_cols), n_rows * n_cols):
    axes.flatten()[idx].set_visible(False)

plt.tight_layout()
plt.show()
```
